# 🧮 Ejercicio 1: Suma de Dígitos (Valor - 5 pts)

### Descripción
Escribe un programa que calcule la suma de los dígitos de un número entero positivo ingresado por el usuario. El programa debe verificar que el número sea positivo utilizando estructuras condicionales. Para el cálculo, debes utilizar un ciclo **while** que extraiga cada dígito del número mediante operaciones de módulo y división entera.

---

### ✅ Salida esperada
El programa debe devolver la suma de todos los dígitos del número ingresado.

**Ejemplo:**
- Entrada: `numero = 1234`  
- Salida: `La suma de los dígitos es: 10`

```python
numero = int(input("Ingresa un número entero positivo: "))

if numero > 0:
    suma = 0
    copia = numero

    while copia > 0:
        digito = copia % 10          
        suma += digito       
        copia = copia // 10             
    print(f"La suma de los dígitos de {numero} es: {suma}")

else:
    print("Error: Debes ingresar un número entero positivo.")
```

# 🔤 Ejercicio 2: Contador de Vocales y Consonantes (Valor - 5 pts)

### Descripción
Escribe un programa que cuente la cantidad de vocales y consonantes en una frase ingresada por el usuario. El programa debe ignorar espacios, números, símbolos y cualquier carácter que no sea una letra del alfabeto. Debes considerar tanto mayúsculas como minúsculas como equivalentes (case-insensitive).

Utiliza un ciclo **for** para iterar sobre cada carácter de la frase y estructuras condicionales para clasificar las letras en vocales o consonantes.

---

### ✅ Salida esperada
El programa debe devolver el conteo de vocales y consonantes por separado.

**Ejemplo:**
- Entrada: `frase = "Hola Mundo!"`  
- Salida: `Vocales: 4, Consonantes: 5`  

```python
frase = input("Escriba una frase: ")

letras_v = "aeiouAEIOU"
letras_c = "bcdfghjklmnñpqrstvwxyzBCDFGHJKLMNÑPQRSTVWXYZ"

vocales = 0
consonantes = 0

for caracter in frase:
    if caracter in letras_v:
        vocales += 1
    elif caracter in letras_c:
        consonantes += 1

print(f"Vocales: {vocales}")
print(f"Consonantes: {consonantes}")
```

# 🔢 Ejercicio 3: Verificación de Número Primo (Valor - 5 pts)

### Descripción
Escribe un programa que determine si un número ingresado por el usuario es primo. 
- Un número primo es aquel que solo es divisible por 1 y por sí mismo. 
- El programa debe utilizar un ciclo for para verificar la divisibilidad del número por todos los enteros menores que él, optimizando el proceso hasta la raíz cuadrada del número.

Incluye estructuras condicionales para manejar casos especiales (números menores o iguales a 1).

---

### ✅ Salida esperada
El programa debe devolver un mensaje claro indicando si el número es primo o no.

**Ejemplo:**
- Entrada: `numero = 7`  
- Salida: `Es primo`  

```python
numero = int(input("Escriba un número entero positivo mayor a 1: "))

if numero > 1:
    es_primo = True

    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            es_primo = False
            print(f"{numero} es divisible por {i}")
            break

    if es_primo:
        print(f"{numero} es primo")
    else:
        print(f"{numero} no es primo")
else:
    print("Error: Debes ingresar un número mayor a 1")
```
